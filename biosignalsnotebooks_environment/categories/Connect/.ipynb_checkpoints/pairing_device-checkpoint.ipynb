For acquiring physiological data there are three systems in interaction: human body (system under exploration); signal acquisition system/sensors (system responsible for collecting data from the human body) and computer/mobile device (system that receives and processes the data collected by the acquisition system).

Plux&#39;s acquisition systems (<a href="https://www.biosignalsplux.com/en/" target="_blank"><span class="color2"><strong>biosignalsplux <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></strong></span></a> and <a href="https://bitalino.com/en/" target="_blank"><span class="color4"><strong>bitalino <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></strong></span></a>) valorize wireless communication between the acquisition system and computer (using <strong><a href="https://www.bluetooth.com/" target="_blank">Bluetooth <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a></strong> technology).

The current <span class="color4"><strong>Jupyter Notebook</strong></span> is intended to explain how Plux&#39;s acquisition systems (<span class="color2"><strong>biosignalsplux</strong></span> in our example) can be quickly connect to a computer in order to ensure future real-time acquisitions through <a href="https://www.biosignalsplux.com/en/software" target="_blank"><span class="color1"><strong>OpenSignals <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></strong></span></a>.
<hr>

<p class="steps">0 - Before starting an acquisition it is mandatory that your Plux acquisition system (in our case <span class="color1"><strong>biosignalsplux</strong></span>) is paired with the computer. It will be described, in the following steps, relevant instructions to fulfill this prerequisite ! </p>
The example is focused on Microsoft Windows 10 Users. For Mac users there are some relevant pages explaining how a Bluetooth device, like <span class="color1"><strong>biosignalsplux</strong></span> can be paired.
<ul>
    <li> <a href="https://support.apple.com/en-us/HT201171">https://support.apple.com/en-us/HT201171</a></li>
    <li> <a href="https://support.apple.com/guide/mac-help/connect-a-bluetooth-device-blth1004/mac">https://support.apple.com/guide/mac-help/connect-a-bluetooth-device-blth1004/mac</a></li>
</ul>

<sub>*It is possible to use the internal Bluetooth of your computer, however to improve the connectivity and have more freedom of movement and avoid communication losses, we strongly recommend to use a Bluetooth Dongle like the one available at <a href="https://store.plux.info/biosignalsplux-accessories/368-bitalino-proven-bluetooth-dongle-820201511.html" target="_blank">Plux Store <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a></sub>.

<p class="steps">1 - Turn on your <span class="color1"><strong>biosignalsplux</strong></span> device</p>

<img src="../../images/other/pair_device_turn_on.gif" width="50%">

<p class="steps">2 - Navigate until the Bluetooth communication system window</p>
<i>This action is achievable through the <strong>"Start Menu"</strong> icon [bottom-left corner of the screen <img src="../../images/record/record_signal_windows_logo.png" style="display:inline;width:20px;margin-top:0px">] >> Settings <img src="../../images/record/record_signal_windows_settings.png" style="display:inline;width:20px;margin-top:0px"> >> Devices <img src="../../images/record/record_signal_windows_devices.png" style="display:inline;width:20px;margin-top:0px"> >> Bluetooth & other devices <img src="../../images/record/record_signal_windows_devices.png" style="display:inline;width:20px;margin-top:0px"></i>

<img src="../../images/other/pair_device_act_bt.gif">

<p class="steps">3 - Enable Bluetooth communication system in your computer</p>

<img src="../../images/other/pair_device_act_bt_on.gif">

<p class="steps">4 - Click on "Add Bluetooth or other device"</p>
<i>We are reaching the pairing stage between our computer and <span class="color1"><strong>biosignalsplux</strong></span> device</i>

<img src="../../images/other/pair_device_bt_add.gif">

<p class="steps">5 - Click on "Bluetooth" option and wait for the list of available devices</p>

<img src="../../images/other/pair_device_bt_option.png">

<p class="steps">6 - Find and click on the entry labeled with "biosignalsplux"</p>

<img src="../../images/other/pair_device_bt_biosignalsplux.png">

<p class="steps">7 - Type the predefined password >> "123"</p>

<img src="../../images/other/pair_device_password.gif">

Mission almost accomplished ! Now you should have your <span class="color1"><strong>biosignalsplux</strong></span> device ready to use in <a href="https://www.biosignalsplux.com/en/software" target="_blank"><span class="color1"><strong>OpenSignals <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></strong></span></a> software.

<img src="../../images/other/pair_device_ready.png">

<p class="steps">8 - Execute <span class="color1"><strong>OpenSignals</strong></span> application in order to get access to the main page</p>

<img src="../../images/open/load_sig_opensignals_main.png">

<p class="steps">9 - Press the <strong>"Device Configuration"</strong> icon for enabling your device</p>

<img src="../../images/other/opensignals_dev.gif">

<p class="steps">10 - Check which is your device mac-address (at the back of your device)</p>

<img src="../../images/other/pair_device_mac.png" width="50%">

<p class="steps">11 - Find your device on the "Device Configuration" page</p>
<i>You should press the "Search" icon in order to the previously added device (through Windows interface) can appear</i>

<img src="../../images/other/opensignals_dev_search.gif">

For finishing the pairing between <span class="color1"><strong>biosignalsplux</strong></span> and the computer, it is only necessary to click on the device box. When a set of configurable options appear the pairing process is officially completed !

<img src="../../images/other/opensignals_dev_click.gif">

The previous instructions represents a reasonable interactive guide for establishing a connection between Plux&#39;s devices and your computer.

If you follow these steps, now you are connected with <span class="color1"><strong>biosignalsplux</strong></span>, which can be the start of an amazing journey !
