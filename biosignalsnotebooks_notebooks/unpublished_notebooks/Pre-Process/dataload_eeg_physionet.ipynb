<strong><span>3 - Load and plot data</span></strong>

<strong><span>3.1 - Importation of the needed packages</span></strong>

```python
# Biosignalsnotebooks python package
import biosignalsnotebooks as bsnb

# Scientific packages
from pyedflib import EdfReader
from numpy import zeros, arange
```

<strong><span>3.2 - Loading of acquired EEG data</span></strong> 
<br>The EEG data of one specific task and one test person is stored in a single edf file, where each row corresponds to an electrode of the 10-20 system. In the following, all electrode data as well as one specific electrode data (O1) is stored and plotted:

<p class="steps"><span class="color2">A - Define the path of the downloaded file (Physionet EEG recording: Task 2 - eyes closed, Test person: 1)</span></p>
For our particular case we defined a relative path of our <span class="color2"><strong>biosignalsnotebooks</strong></span> project.

```python
#define relative file path for Physionet EEG recording: Task 2 - eyes closed, Test person: 1
relative_file_path = '../../signal_samples/S001R02.edf'

#define the row of the electrode no.
electrode_row = 60
```

<p class="steps"><span class="color4">B - Read the .edf file and prepare the data structure "data" where the file content will be stored</span></p>

```python
# Read the edf file and store the data       
f = EdfReader(relative_file_path)
n = f.signals_in_file
signal_labels =  f.getSignalLabels()

# Prepare the data structure responsible for storing the .edf file content.
data = np.zeros((n,f.getNSamples()[0]))
```

<p class="steps"><span class="color7">C - Read the signal stored in each row of data</span></p>

```python
#read the signal of all rows
for i in np.arange(n):
     data[i, :] = f.readSignal(i)
```

<p class="steps"><span class="color13">D - Storage of specific data and execution of the needed procedures to close our .edf file</span></p>

```python
# store specific electrode data
data_electrode = data[electrode_row]

#store all data
data_eeg = data 

#close the edf file
f._close() 
del f
```

The plot illustrates the raw data of all electrodes according to the 10-20 system for the Baseline Test 2 (60 seconds eyes closed - Test person 1):

```python
print(type(list(map(list, data_eeg))[0][0]))
bsnb.plot(list(map(list, data_eeg))[0])
```

```python
# delete this code and use bsnb.plot
import matplotlib.pyplot as plt
#plotting all electrodes: 
plt.figure(figsize=(20, 5)) 
plt.plot(data_eeg)
plt.title('RAW EEG DATA - Test Person 1 - Baseline Test 2 - All Electrodes')
plt.xlabel('time [sec]')
plt.ylabel('[uV]')
plt.show()
```

The time axis is generated for the plot of the single electrode:

```python
#generate time axis for single electrode plot
sr=160 #sampling rate PhysioNet
time = bsnb.generate_time(data_electrode, sr)
```

The Figure shows the plot of a single electrode (O1) for the above mentioned Baseline test:

```python
#plot of EEG Data of single electrode
bsnb.plot(time,data_electrode, y_axis_label = "[uV]",x_axis_label = "time [sec.]",title='RAW EEG DATA - Electrode: O1', x_range=(0,61), grid_lines=1, grid_columns=1,grid_plot=True)
```
