The acquired electrophysiological signals have always two intrinsic components.
The signal we really want to acquire/study and noise, i.e. the acquisition component that is not relevant for the study purposes.

Noise can have different origins, such as in random events or due to voluntary/involuntary movements of the subject under analysis that affect the signal acquisition <a href="https://en.wikipedia.org/wiki/Noise_(signal_processing)" target="_blank"><img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a>.

So, filtering <a href="https://en.wikipedia.org/wiki/Filter_(signal_processing)" target="_blank"><img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a> is a fundamental step that needs to be applied to the signal, in order to ensure the maximisation of signal to noise ratio <a href="https://en.wikipedia.org/wiki/Signal-to-noise_ratio" target="_blank"><img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a>. Filtering can be achieved by hardware, having the analogical systems a great relevance, or by software using digital filters.

In this **<span class="color5">Jupyter Notebook</span>** it will be demonstrated how to use median filter for the signal<a href="https://en.wikipedia.org/wiki/Median_filter" target="_blank"><img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a>. Median filtering is the popular method of noise removal, employed in applications involving speech, signal and image processing. This nonlinear technique has proven to be a good alternative linear filtering as it can effectively supress impusle noise while preserving edge information edge information.

<p class="steps">2 - Apply Median filter to signal </p> 

```python
# Data loading
t = arange(0, 1, 0.02)
s = sin(2 * pi * t)
s[ [8,20,25] ] = -0.9
ns = signal.medfilt(s, 5)


plt.plot(t, s, "-r", lw=2, ms=11, alpha=.2)
plt.plot(t, ns, "ob", lw=2, ms=6)
plt.legend(["noisy signal", "median filter"])
plt.show()
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy import signal

#x1= np.loadtxt("https://drive.google.com/open?id=1sXJy1h5schmk22PaWcYHp9KrzPBbkWNS")
data, header = bsnb.load_signal("https://drive.google.com/open?id=1sXJy1h5schmk22PaWcYHp9KrzPBbkWNS", get_header=True)
x1 = data["00:07:80:79:6F:DB"]["CH1"]
x2 = x1 + np.random.rand(len(x1)) * sampling_rate # add noise to the signal

sampling_rate = header["00:07:80:79:6F:DB"]["sampling rate"]
t = np.linspace(0,len(x1)/sampling_rate,len(x1)) # create a time signal

y1 = sp.signal.medfilt(x2,sampling_rate+1) # add noise to the signal

# plot the results
plt.subplot(3,1,1)
plt.plot(t,x2,'y-')
plt.title('input wave')
plt.xlabel('time')
plt.subplot(3,1,2)
plt.plot(t,y1,'y-')
plt.title('filtered wave')
plt.xlabel('time')
plt.subplot(3,1,3)
plt.plot(t,x2,'y-', t, y1, 'b-')
plt.title('combined plots')
plt.xlabel('time')
plt.show()
```
